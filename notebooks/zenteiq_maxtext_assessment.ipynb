```python
# Cell 1: Clone MaxText from GitHub and install dependencies

import subprocess
import sys
import os

# ── Detect backend ──────────────────────────────────────────────
def detect_backend():
    try:
        result = subprocess.run(["nvidia-smi"], capture_output=True)
        if result.returncode == 0:
            return "gpu"
    except FileNotFoundError:
        pass
    try:
        with open("/sys/devices/virtual/dmi/id/product_name") as f:
            if "Google" in f.read():
                import requests
                r = requests.get("http://metadata.google.internal/computeMetadata/v1/instance/attributes/",
                                 headers={"Metadata-Flavor": "Google"}, timeout=2)
                if "tpu" in r.text.lower():
                    return "tpu"
    except:
        pass
    return "cpu"

BACKEND = detect_backend()
print(f"✅ Detected backend: {BACKEND.upper()}")

# ── Clone MaxText repo ───────────────────────────────────────────
if not os.path.exists("maxtext"):
    print("\n📦 Cloning MaxText repository...")
    subprocess.run(["git", "clone", "https://github.com/google/maxtext.git"], check=True)
    print("✅ Cloned successfully!")
else:
    print("✅ MaxText directory already exists, skipping clone.")

os.chdir("maxtext")
print(f"📁 Working directory: {os.getcwd()}")

# ── Run MaxText setup script ─────────────────────────────────────
print("\n⚙️ Running MaxText setup (bash setup.sh)...")
result = subprocess.run(
    ["bash", "setup.sh"],
    capture_output=True, text=True
)
print(result.stdout[-3000:] if len(result.stdout) > 3000 else result.stdout)
if result.returncode != 0:
    print("⚠️ STDERR:", result.stderr[-2000:])
else:
    print("✅ Setup complete!")
```

```python
# Cell 3 (GPU): Scan all missing modules then install all at once

import subprocess
import sys
import re
import json
import os

os.chdir("/content/maxtext")

# ── Step 1: Scan for ALL missing imports ─────────────────────────
print("🔍 Scanning for ALL missing imports...")
result = subprocess.run(
    [sys.executable, "-c", "from maxtext.trainers.pre_train import train"],
    capture_output=True, text=True, cwd="/content/maxtext"
)
missing = []
for line in result.stderr.split("\n"):
    if "ModuleNotFoundError" in line or "ImportError" in line:
        print(line)
        # Extract module name
        m = re.search(r"No module named '([^']+)'", line)
        if m:
            missing.append(m.group(1).split(".")[0])

print(f"\n📦 Missing modules: {list(set(missing))}")

# ── Step 2: Install known mappings ───────────────────────────────
module_to_package = {
    "tokamax": "tokamax",
    "drjax": "drjax",
    "grain": "grain-nightly",
    "tensorflow": "tensorflow-cpu",
    "tensorstore": "tensorstore",
    "etils": "etils",
    "clu": "clu",
    "flax": "flax>=0.12.7",
    "optax": "optax>=0.2.8",
    "orbax": "orbax-checkpoint>=0.11.39",
    "aqt": "aqtp",
    "qwix": "qwix",
    "ml_goodput_measurement": "ml-goodput-measurement",
    "ml_collections": "ml_collections",
    "pathwaysutils": "pathwaysutils",
    "sentencepiece": "sentencepiece",
    "transformers": "transformers",
}

packages_to_install = []
for mod in set(missing):
    pkg = module_to_package.get(mod, mod)
    packages_to_install.append(pkg)

# Always install tokamax explicitly
extra = ["tokamax", "clu", "etils[array_types]", "tensorstore"]
all_pkgs = list(set(packages_to_install + extra))

print(f"\n📦 Installing: {all_pkgs}")
result2 = subprocess.run(
    [sys.executable, "-m", "pip", "install"] + all_pkgs + ["-q"],
    capture_output=True, text=True
)
if result2.returncode != 0:
    print("⚠️", result2.stderr[-1000:])
else:
    print("✅ All installed!")

# ── Step 3: Re-scan to confirm clean ─────────────────────────────
print("\n🔍 Re-scanning after install...")
result3 = subprocess.run(
    [sys.executable, "-c", "from maxtext.trainers.pre_train import train; print('✅ All imports OK')"],
    capture_output=True, text=True, cwd="/content/maxtext"
)
print(result3.stdout)
remaining = [l for l in result3.stderr.split("\n") if "ModuleNotFoundError" in l or "ImportError" in l]
if remaining:
    print("Still missing:")
    for l in remaining:
        print(l)
else:
    print("✅ No missing modules!")

# ── Step 4: Run training ──────────────────────────────────────────
cmd = [
    "python3", "-m", "maxtext.trainers.pre_train.train",
    "src/maxtext/configs/base.yml",
    "model_name=qwen3-0.6b",
    "base_output_directory=/content/maxtext_outputs",
    "run_name=qwen_0.6b_gpu_run",
    "dataset_type=synthetic",
    "steps=50",
    "per_device_batch_size=1",
    "max_target_length=512",
    "enable_checkpointing=false",
    "async_checkpointing=false",
    "attention=dot_product",
    "dtype=bfloat16",
    "weight_dtype=bfloat16",
    "tokenizer_type=huggingface",
    "tokenizer_path=Qwen/Qwen3-0.6B",
    "scan_layers=false",
    "remat_policy=minimal",
    "log_period=1",
]

print("\n🚀 Starting Qwen 0.6B on GPU (50 steps)...")
result4 = subprocess.run(cmd, capture_output=True, text=True, cwd="/content/maxtext")

with open("/content/qwen_0.6b_gpu_full.log", "w") as f:
    f.write(result4.stderr)

param_match = re.search(r"number parameters:\s*([\d.]+)\s*billion", result4.stderr)
if param_match:
    print(f"📊 Model size: {param_match.group(1)}B parameters")

pattern = r"completed step:\s*(\d+), seconds:\s*([\d.]+), TFLOP/s/device:\s*([\d.]+), Tokens/s/device:\s*([\d.]+), total_weights:\s*(\d+), loss:\s*([\d.]+), lm_loss:\s*([\d.]+), perplexity:\s*([\d.]+)"
matches = re.findall(pattern, result4.stderr)

print(f"\n✅ Captured {len(matches)} steps\n")
print(f"{'Step':>5} {'Seconds':>10} {'TFLOP/s':>10} {'Tok/s':>10} {'Loss':>10} {'Perplexity':>12}")
print("-" * 65)
for m in matches:
    step, sec, tflops, toks, weights, loss, lm_loss, ppl = m
    print(f"{step:>5} {sec:>10} {tflops:>10} {toks:>10} {loss:>10} {ppl:>12}")

if result4.returncode != 0 and len(matches) == 0:
    print("\n=== STDERR (last 2000) ===")
    print(result4.stderr[-2000:])

metrics = [
    {"step": int(m[0]), "seconds": float(m[1]), "tflops_per_device": float(m[2]),
     "tokens_per_sec": float(m[3]), "loss": float(m[5]), "lm_loss": float(m[6]),
     "perplexity": float(m[7])}
    for m in matches
]
with open("/content/qwen_0.6b_gpu_metrics.json", "w") as f:
    json.dump({"model": "qwen3-0.6b", "backend": "gpu", "steps": metrics}, f, indent=2)

print(f"\nReturn code: {result4.returncode}")
```

```python
# Cell 5 (GPU): Re-parse full log to extract all 50 steps correctly

import re
import json

def parse_log(log_path, model, backend):
    with open(log_path) as f:
        content = f.read()

    pattern = r"completed step:\s*(\d+), seconds:\s*([\d.]+), TFLOP/s/device:\s*([\d.]+), Tokens/s/device:\s*([\d.]+), total_weights:\s*(\d+), loss:\s*([\d.]+), lm_loss:\s*([\d.]+), perplexity:\s*([\S]+)"
    matches = re.findall(pattern, content)

    print(f"\n📊 {model} on {backend} — {len(matches)} steps captured")
    print(f"\n{'Step':>5} {'Seconds':>10} {'TFLOP/s':>10} {'Tok/s':>10} {'Loss':>12} {'Perplexity':>20}")
    print("-" * 75)

    metrics = []
    for m in matches:
        step, sec, tflops, toks, weights, loss, lm_loss, ppl = m
        print(f"{step:>5} {sec:>10} {tflops:>10} {toks:>10} {loss:>12} {ppl:>20}")
        metrics.append({
            "step": int(step),
            "seconds": float(sec),
            "tflops_per_device": float(tflops),
            "tokens_per_sec": float(toks),
            "total_weights": int(weights),
            "loss": float(loss),
            "lm_loss": float(lm_loss),
            "perplexity": ppl   # keep as string since some are 'inf'
        })
    return metrics

# ── Parse Qwen 0.6B GPU ───────────────────────────────────────────
metrics_0_6b = parse_log(
    "/content/qwen_0.6b_gpu_full.log",
    "Qwen 0.6B", "GPU"
)
with open("/content/qwen_0.6b_gpu_metrics.json", "w") as f:
    json.dump({"model": "qwen3-0.6b", "backend": "gpu", "steps": metrics_0_6b}, f, indent=2)
print(f"\n✅ Saved qwen_0.6b_gpu_metrics.json ({len(metrics_0_6b)} steps)")

# ── Also re-parse CPU logs ────────────────────────────────────────
metrics_0_6b_cpu = parse_log(
    "/content/qwen_0.6b_cpu_full.log" if __import__('os').path.exists("/content/qwen_0.6b_cpu_full.log") else "/content/qwen_0.6b_cpu_full.log",
    "Qwen 0.6B", "CPU"
) if __import__('os').path.exists("/content/qwen_0.6b_cpu_full.log") else []
```

```python
import subprocess
import re
import json
import os

os.chdir("/content/maxtext")
os.makedirs("/content/maxtext_outputs", exist_ok=True)

# deepseek-custom is our base (already a small test config derived from 671B)
# We override to strip experimental features (indexer, engram, mhc)
# and tighten dims to ensure clean single-device run under 1B params.
#
# MoE parameter count (approximate):
#   Embedding:  vocab(129280) x emb(512) = ~66M
#   Per layer:  attention(MLA) + dense_ffn(first layer) + MoE FFN(remaining layers)
#   MoE layer:  num_experts(8) x moe_mlp_dim(512) x emb(512) x 2 = ~4M per layer
#   5 MoE layers x ~4M + 1 dense layer + embeddings \u2248 ~200-400M total
#   With shared_expert + routing overhead \u2248 safely under 1B

cmd = [
    "python3", "-m", "maxtext.trainers.pre_train.train",
    "src/maxtext/configs/base.yml",
    "model_name=deepseek-custom",
    "override_model_config=True", # Add this line to allow overriding model config parameters
    # Architecture overrides \u2014 scale DOWN from deepseek-custom defaults
    "base_emb_dim=512",
    "base_num_query_heads=8",
    "base_num_kv_heads=8",
    "base_mlp_dim=1024",
    "base_moe_mlp_dim=512",
    "base_num_decoder_layers=6",
    "first_num_dense_layers=1",
    "num_experts=8",
    "num_experts_per_tok=2",
    "shared_experts=1",
    # MLA attention dims \u2014 keep small
    "q_lora_rank=128",
    "kv_lora_rank=64",
    "qk_nope_head_dim=32",
    "qk_rope_head_dim=16",
    "v_head_dim=64",
    # Disable all experimental features that need multi-device or special deps
    "use_indexer=false",
    # "mhc_expansion_rate=0", # Removed due to validation error: "Input should be greater than 0"
    "engram_layers=[]",
    # RoPE \u2014 simplified for single device
    "max_position_embeddings=4096",
    "original_max_position_embeddings=4096",
    "rope_factor=1",
    # Training
    "base_output_directory=/content/maxtext_outputs",
    "run_name=deepseek_moe_tpu_run",
    "dataset_type=synthetic",
    "steps=50",
    "per_device_batch_size=1",
    "max_target_length=512",
    "enable_checkpointing=false",
    "async_checkpointing=false",
    "attention=dot_product",
    "dtype=bfloat16",
    "weight_dtype=bfloat16",
    "tokenizer_type=huggingface",
    "tokenizer_path=Qwen/Qwen3-0.6B",   # vocab size mismatch is fine for synthetic
    "scan_layers=false",
    "remat_policy=minimal",
    "log_period=1",
]

print("🚀 Starting DeepSeek MoE (<1B) on TPU (50 steps)...")
result = subprocess.run(cmd, capture_output=True, text=True, cwd="/content/maxtext")

with open("/content/deepseek_moe_tpu_full.log", "w") as f:
    f.write(result.stderr)

param_match = re.search(r"number parameters:\s*([\d.]+)\s*(billion|million)", result.stderr)
if param_match:
    print(f"📊 Model size: {param_match.group(1)} {param_match.group(2)} parameters")

pattern = r"completed step:\s*(\d+), seconds:\s*([\d.]+), TFLOP/s/device:\s*([\d.]+), Tokens/s/device:\s*([\d.]+), total_weights:\s*(\d+), loss:\s*([\d.]+), lm_loss:\s*([\d.]+), perplexity:\s*([\S]+)"
matches = re.findall(pattern, result.stderr)

print(f"\n✅ Captured {len(matches)} steps\n")
print(f"{'Step':>5} {'Seconds':>10} {'TFLOP/s':>10} {'Tok/s':>10} {'Loss':>12} {'Perplexity':>20}")
print("-" * 75)
for m in matches:
    step, sec, tflops, toks, weights, loss, lm_loss, ppl = m
    print(f"{step:>5} {sec:>10} {tflops:>10} {toks:>10} {loss:>12} {ppl:>20}")

if result.returncode != 0 and len(matches) == 0:
    print("\n=== STDERR (last 2500) ===")
    print(result.stderr[-2500:])

metrics = [
    {"step": int(m[0]), "seconds": float(m[1]), "tflops_per_device": float(m[2]),
     "tokens_per_sec": float(m[3]), "total_weights": int(m[4]),
     "loss": float(m[5]), "lm_loss": float(m[6]), "perplexity": m[7]}
    for m in matches
]
with open("/content/deepseek_moe_tpu_metrics.json", "w") as f:
    json.dump({"model": "deepseek-custom-sub1b", "backend": "tpu", "steps": metrics}, f, indent=2)

print(f"\n✅ Saved to /content/deepseek_moe_tpu_metrics.json")
print(f"Return code: {result.returncode}")
```

```python
import subprocess
import os
import shutil

# Define variables for GitHub access (already defined in previous cells, but good for clarity)
GITHUB_USERNAME = "nageswarao7"
GITHUB_TOKEN    = "ghp_YOUR_GITHUB_TOKEN_PLACEHOLDER"
REPO_NAME       = "zentiq-maxtext-assignment"
REPO_URL = f"https://{GITHUB_USERNAME}:{GITHUB_TOKEN}@github.com/{GITHUB_USERNAME}/{REPO_NAME}.git"
YOUR_EMAIL      = "nageswararaovutla7@gmail.com"
YOUR_NAME       = "Nageswara Rao Vutla"

# Ensure we are in the root /content directory before attempting git operations
os.chdir("/content")

# Clone the repository fresh to ensure we have the latest state
if os.path.exists("/content/submission"):
    shutil.rmtree("/content/submission")

clone_result = subprocess.run(["git", "clone", REPO_URL, "/content/submission"],
                               capture_output=True, text=True)

if clone_result.returncode != 0:
    print(f"⚠️ Git clone failed: {clone_result.stderr}")
    # Ensure the directory still exists to write files, even if clone failed
    os.makedirs("/content/submission", exist_ok=True)
else:
    print("✅ Git clone successful.")

# Ensure the configs directory exists
os.makedirs("/content/submission/configs", exist_ok=True)

# Copy deepseek-custom.yml to the submission configs directory
src_path = "/content/maxtext/src/maxtext/configs/models/deepseek-custom.yml"
dst_path = "/content/submission/configs/deepseek-custom.yml"

if os.path.exists(src_path):
    shutil.copy(src_path, dst_path)
    print(f"✅ Copied {os.path.basename(src_path)} to submission directory.")
else:
    print(f"⚠️  Source file not found: {src_path}")

# Change to the submission directory for git operations
os.chdir("/content/submission")

# Git commands to add, commit, and push the file
cmds = [
    ["git", "config", "user.email", YOUR_EMAIL],
    ["git", "config", "user.name", YOUR_NAME],
    ["git", "add", "configs/deepseek-custom.yml"],
    ["git", "commit", "-m", "Add deepseek-custom.yml config file"],
    ["git", "push", "origin", "main"]
]

for cmd in cmds:
    result = subprocess.run(cmd, capture_output=True, text=True)
    display_cmd = " ".join(cmd).replace(GITHUB_TOKEN, "***")
    if result.returncode != 0 and "nothing to commit" not in result.stderr:
        print(f"⚠️ Failed: {display_cmd}")
        print(result.stderr[-500:])
    else:
        print(f"✅ {display_cmd}")

print(f"\n🎉 deepseek-custom.yml pushed! View at: https://github.com/{GITHUB_USERNAME}/{REPO_NAME}")
```
